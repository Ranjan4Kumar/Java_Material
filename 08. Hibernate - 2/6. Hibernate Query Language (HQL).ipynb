<h1 class="entry-title" itemprop="headline">Hibernate Query Language (HQL)

</h1>

<h2>HQL:</h2>

<p>HQL stands for Hibernate Query Language. HQL syntax is quite similar to SQL syntax but it performs operations on objects and properties of persistent classes instead of tables and columns. Hibernate framework translate HQL queries into database specific queries to perform action.</p>

<p>Query interface provides the methods and functionality to represent and manipulate a HQL query in the object oriented way.<br> <strong>Note: We can directly use SQL statements in hibernate.</strong></p>
