<h1 class="entry-title" itemprop="headline">Hibernate framework</h1>

<p>Hibernate framework is an object-relational mapping library for the Java language. It is an open source persistent framework started by Gavin King in 2001. It provides the facility to map the java classes to relational database tables and Java data types to SQL data types.<br> <strong>Note: Hibernate framework acts as an intermediate layer between java objects and relational database.</strong></p>

<h2>Advantages of Hibernate framework:</h2>

<ol><li>Hibernate framework is an open source framework.</li><li>Hibernate framework is a high performance framework because of its internal cache.</li><li>It maps java classes to database tables using XML files and if any change is made to the database than we have to change these XML files only.</li><li>It provides a database independent query language HQL (Hibernate Query Language).</li><li>It provides the facility to create tables from your java application.</li><li>No application server is required.</li></ol>
